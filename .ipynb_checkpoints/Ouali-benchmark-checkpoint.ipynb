```python
import pandas as pd
import datacompy
benchmark_file = 'input/benchmark-nomsATC.tsv'
ouali_files = ['input/AI-P1-aligned.tsv', 'input/AI-P2-aligned.tsv', 'input/AI-P3-aligned.tsv']
rero_files = ['input/RERO-aligned-1.txt', 'input/RERO-aligned-2.txt']
```

```python
benchmark_data = pd.read_csv(benchmark_file, sep='\t')
ouali_data = pd.concat((pd.read_csv(f, sep='\t') for f in ouali_files))
rero_data = pd.concat((pd.read_csv(f, sep='\t', encoding = "ISO-8859-1") for f in rero_files))
```

```python
bench_no_align = benchmark_data[benchmark_data['nombre de candidats'] == 0]
bench_align = benchmark_data[benchmark_data['nombre de candidats'] != 0]

print('Alignements: ' + str(len(bench_align)))
print('Non-alignements: ' + str(len(bench_no_align)))
```

```python
ouali_align = ouali_data.query('`nombre de candidats` != 0 & `décision d\'alignement` == "auto"')
ouali_no_align = ouali_data.query('`nombre de candidats` == 0 & `décision d\'alignement` == "auto"')

print('Alignements: ' + str(len(ouali_align)))
print('Non-alignements: ' + str(len(ouali_no_align)))
```

```python
compare = datacompy.Compare(bench_align, ouali_align, join_columns=['id source', 'id cible'], 
                            df1_name='RERO Benchmark', df2_name='Ouali')
print(compare.report())
```

```python
common_align = pd.merge(bench_align, ouali_align, on=['id source', 'id cible'], how='inner')
```
